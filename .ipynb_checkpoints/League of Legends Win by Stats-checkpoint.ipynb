```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt

data = pd.read_csv("high_diamond_ranked_10min.csv", index_col="gameId")
data.head()
```

Let´s start investigating the data. One would expect that the game wins should be roughly symetrical, meaning that the percentage of wins of red-side and blue-side should both be around 50%.

```python
print("Blue side victory percentage: " + str( np.round(data[data["blueWins"] == 1].shape[0]/data.shape[0]*100, 3)) + "%")
print("Red side victory percentage: " + str( np.round(data[data["blueWins"] == 0].shape[0]/data.shape[0]*100, 3)) + "%")
```

As we can see this is holds farily well, which is expected. This means that most of the categories are not important in absolute terms. Instead what is important is the relative difference in both teams.

Let´s reshape the data to reflect this fact. Let´s first reduce the number of columns by recognizing some categorical attributes. This is: **blueFirstBlood/redFirstBlood**. 

We drop the columns that refer to the red team.

```python
data = data.drop(["redFirstBlood"], axis=1)
```

Some of the data is reperative. For instance we have **blueGoldDiff** and **redGoldDiff**, but the relationship between them is clear. Similarly to this, **blueKills** and **redDeaths** has a similar inverse relationship. We drop these repetative columns.

```python
data = data.drop([
    "redGoldDiff", 
    "redExperienceDiff",
    "redDeaths",
    "blueDeaths",
    "blueGoldPerMin", 
    "redGoldPerMin", 
    "blueTotalMinionsKilled", 
    "redTotalMinionsKilled"],
    axis=1)
```

Let´s look at how the data is distributed and what is the importance of each attribute to the percentage of winning the game.

```python
data.corr()["blueWins"].sort_values(ascending=False)
```

Quite expectedly the main contributors to a team winning seem to be if a team is ahead in gold and/or experience as this leads to item purchases and inherent champion level scaling.

Let´s look at what was the average gold and experience difference between BLUE and RED team in blue team wins and losses.

```python
fig, ax = plt.subplots(1, 2, figsize=(15,12))
ax[0].set_title("Gold difference in wins and losses")
ax[1].set_title("Experience difference in wins and losses")

# Gold difference
ax[0].hist([
    data[data["blueWins"]==0]["blueGoldDiff"], 
    data[data["blueWins"]==1]["blueGoldDiff"]], 
    bins=50, color=["red", "black"], density=True)

# Exp difference
ax[1].hist([
    data[data["blueWins"]==0]["blueExperienceDiff"], 
    data[data["blueWins"]==1]["blueExperienceDiff"]],
    bins=50, color=["red", "black"], density=True)

ax[0].legend(["Loss", "Win"], fontsize ="xx-large")
ax[1].legend(["Loss", "Win"], fontsize ="xx-large")
plt.show()
```

We can see that some teams even with an experience and gold deficit at 10 min (left side of distribution) still managed to claw themselves back into the game. This already makes us think that our Machine Learning model will have trouble achieving a very good accuracy as 10 minutes of gameplay may simply be too little to reach any reliable conclusions.

```python
fig, ax = plt.subplots(figsize=(15,12))
ax.set_title("2D plot of Exp. & Gold difference")

ax.scatter(
    x=data[data["blueWins"]==0]["blueGoldDiff"]/data[data["blueWins"]==0]["blueTotalGold"],
    y=data[data["blueWins"]==0]["blueExperienceDiff"]/data[data["blueWins"]==0]["blueTotalExperience"],    
    color="red")

ax.scatter(
    x=data[data["blueWins"]==1]["blueGoldDiff"]/data[data["blueWins"]==1]["blueTotalGold"],
    y=data[data["blueWins"]==1]["blueExperienceDiff"]/data[data["blueWins"]==1]["blueTotalExperience"],    
    color="black")


plt.xlabel('Relative Gold Diff.', fontsize=18)
plt.ylabel('Relative Experience Diff.', fontsize=18)
ax.legend(["Loss", "Win"], fontsize ="xx-large")
plt.show()
```

We can see that in blue wins and losses the distribution is squeezed in the experience difference category compared to gold difference.

Let´s now look at the importance of taking the dragon and the herald.

```python
fig, ax = plt.subplots(1, 2, figsize=(15,12))
ax[0].set_title("Dragon takes")
ax[1].set_title("Herald takes")

# Gold difference
ax[0].hist([
    data[data["blueWins"]==0]["blueDragons"], 
    data[data["blueWins"]==1]["blueDragons"]], 
    color=["red", "black"], histtype='bar')

# Exp difference
ax[1].hist([
    data[data["blueWins"]==0]["blueHeralds"], 
    data[data["blueWins"]==1]["blueHeralds"]], 
    color=["red", "black"], histtype='bar')

ax[0].legend(["Loss", "Win"], fontsize ="xx-large")
ax[1].legend(["Loss", "Win"], fontsize ="xx-large")
plt.show()
```

The data suggests seems to be assymetrical and therefore suggests that it is not as important to take the drake or heralds, as long as you can prevent the other team from taking it. It seems that the heralds is often not taken before 10 minutes, therefore we analyze the only the drake data at this point of time.

Let´s compare the importance of the drake only when it was taken in the game.

```python
data_drake_taken = data[(data["blueDragons"] == 1) | (data["redDragons"] == 1)]
print("Percent of drake taken: " + str(np.round(data_drake_taken.shape[0]/data.shape[0]*100, 2)) + "%")

fig, ax = plt.subplots(figsize=(15,12))
ax.set_title("Dragon takes")

# Gold difference
ax.hist([
    data_drake_taken[data_drake_taken["blueWins"]==0]["blueDragons"], 
    data_drake_taken[data_drake_taken["blueWins"]==1]["blueDragons"]], 
    color=["red", "black"], histtype='bar')

ax.legend(["Loss", "Win"], fontsize ="xx-large")
plt.show()
```

This now makes much more sense. We can see that if a team takes a drake (in the case any team gets a drake at all) it has a significant probability of winning the game.

To more properly take this into account we introduce a new attribute instead of the old drake attributes.

**blueDrakeDiff** = **blueDragons** - **redDragons**

If no team takes the drake than its value is 0. If blue team takes it its value is 1, else it is -1.

```python
data["blueDrakeDiff"] = data["blueDragons"] - data["redDragons"]
```
